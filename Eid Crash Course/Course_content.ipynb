# What is ML?
Machine Learning or simply ML is the measures taken to achieve AI.

> AI capabilities include adapting to new environment, reasoning ability and prblem solving with available resources. If a machine has all these of these capabilities we call it artificially intelligent. Machine learning is used to achive all of these or a subset of these capabilities. For this purpose we train the machine with relevent data.

A model is just an object generated by a machine lerning algorithm through the training.

We apply a machine learning algorithm on the data.

The more data we have the chances of our model to be more accurate. Data often requires pre processing and tweeks before fed to an algorithm.

In ML we deal with certain types of problems. These include -->
- Classification
- Regression
- Clustering
- Dimensionality reduction

To deal with these scenerios we apply various ML algorithm. Based on these algorithms machine learning can be broadly classified into -->
- Supervised learning
- Unsupervised learning
- Reinforced learning
- Semi-supervised learning

> Up until now we've concentrated on supervised learning and we're only dealing with classification problems.
