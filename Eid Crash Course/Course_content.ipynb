# What is ML?
Machine Learning or simply ML is the measures taken to achieve AI.

> AI capabilities include adapting to new environment, reasoning ability and prblem solving with available resources. If a machine has all these of these capabilities we call it artificially intelligent. Machine learning is used to achive all of these or a subset of these capabilities. For this purpose we train the machine with relevent data.

A model is just an object generated by a machine lerning algorithm through the training.

We apply a machine learning algorithm on the data.

The more data we have the chances of our model to be more accurate. Data often requires pre processing and tweeks before fed to an algorithm.

In ML we deal with certain types of problems. These include -->
- Classification
- Regression
- Clustering
- Dimensionality reduction

To deal with these scenerios we apply various ML algorithm. Based on these algorithms machine learning can be broadly classified into -->
- Supervised learning
- Unsupervised learning
- Reinforced learning
- Semi-supervised learning

> Up until now we've concentrated on supervised learning and we're only dealing with classification problems.

```python
import pandas as pd
df = pd.read_csv('../predictive_maintenance.csv')
df.head()
```

```python
y = df['Target'].values

X = df[['Air_temperature',
        'Process_temperature','Rotational_speed',
        'Torque','Tool_wear']].values
```

```python
print(X[:5])
```

# Evaluation Metrics
Evaluation metric are various measures to better understand how our model is performing or expected to perform after it's been trained.
> Accuracy is the most common evaluation measure.

Accuary does not provide a proper insight of the model performance if the dataset we're using is imbalanced or biased. Hence we use several other metrics to better gain insight of our model. These are -->
- Precision
- Recall
- F1 score

```python
# necessary imports
from sklearn.metrics import accuracy_score as acc
from sklearn.metrics import precision_score as prec
from sklearn.metrics import recall_score as rec
from sklearn.metrics import f1_score as f1
from sklearn.metrics import confusion_matrix as cfm
```

```python
from sklearn.linear_model import LogisticRegression

modelLgR = LogisticRegression()

modelLgR.fit(X,y)

print(modelLgR.coef_,modelLgR.intercept_)
```

```python
modelLgR.predict([[298.9,309.3,1581,30.7,65]])
```

```python
y_predict = modelLgR.predict(X)
print(y_predict[:10])
```

```python
print((y == y_predict).sum())
```

```python
print(acc(y,y_predict))
```

```python
print(prec(y,y_predict))
```

```python
print(rec(y,y_predict))
```

```python
print(f1(y,y_predict))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8)
```

```python
print(X_test.shape)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)


print(acc(y_test,y_pred))
print(prec(y_test,y_pred))
```

```python
y_pred_proba = model.predict_proba(X_train)

print(y_pred_proba)
```

```python
threshold = 0.75

y_pred2 = model.predict_proba(X_test)[:,1] > threshold
```

```python
print(prec(y_test,y_pred2))
print(rec(y_test,y_pred2))
```
